## Background

The SAT and ACT are standardized tests that many colleges and universities in the United States require for their admissions process. This score is used along with other materials such as grade point average (GPA) and essay responses to determine whether or not a potential student will be accepted to the university.

The SAT has two sections of the test: Evidence-Based Reading and Writing and Math ([*source*](https://www.princetonreview.com/college/sat-sections)). The ACT has 4 sections: English, Mathematics, Reading, and Science, with an additional optional writing section ([*source*](https://www.act.org/content/act/en/products-and-services/the-act/scores/understanding-your-scores.html)). They have different score ranges, which you can read more about on their websites:
* [SAT](https://collegereadiness.collegeboard.org/sat)
* [ACT](https://www.act.org/content/act/en.html)

Standardized tests have long been a controversial topic for students, administrators, and legislators. Since the 1940's, an increasing number of colleges have been using scores from sudents' performances on tests like the SAT and the ACT as a measure for college readiness and aptitude ([*source*](https://www.minotdailynews.com/news/local-news/2017/04/a-brief-history-of-the-sat-and-act/)). Supporters of these tests argue that these scores can be used as an objective measure to determine college admittance. Opponents of these tests claim that these tests are not accurate measures of students potential or ability and serve as an inequitable barrier to entry. Lately, more and more schools are opting to drop the SAT/ACT requirement for their Fall 2021 applications ([*read more about this here*](https://www.cnn.com/2020/04/14/us/coronavirus-colleges-sat-act-test-trnd/index.html)).

The U.S. Department of Education is the agency of the federal government that establishes policies for education. In 2017, 2.03 million high school students took the ACT <sup> 1 </sup>, and 1.8 million students took the SAT <sup> 2 </sup>. By studying the relationship between participation rates and test scores, it can give recommendations on whether mandating students to partcipate in ACT/SAT is beneficial in raising education standards in the country.

```python
def mean_calc(list_1):
    return sum(list_1)/len(list_1)

mean_calc([10,20,30])
```

2. Manually calculate standard deviation:

    The formula for standard deviation is below:

    $$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

    Where $x_i$ represents each value in the dataset, $\mu$ represents the mean of all values in the dataset and $n$ represents the number of values in the dataset.

    Write a function that takes in values and returns the standard deviation of the values using the formula above. Hint: use the function you wrote above to calculate the mean! Use the list of numbers you created above to test on your function.
    
    *Note*: Do not use any standard deviation methods built-in to any Python libraries to do this! This should be done without importing any additional libraries.

```python
def sigma(list_1):
    
    total = 0
    for i in list_1:
        
        total+=(i - mean_calc(list_1))**2
    
    final = np.sqrt(total/len(list_1))
    
    return final

sigma(pd.Series([10,20,30,40,50,60]))
```

Through the histogram plots, we suspect that the participation rates for SAT and ACT follows a bimodal distribution. However, it is not too obvious from a histogram plot. Let us plot a distplot with a kernel density estimate (KDE) in order to better verify our analysis.

```python
def dist_plot(df,list_of_inputs,y_label_input,title_input):
    
    """Plotting of distplot"""
    
    rows = int(np.ceil(len(list_of_inputs)/2))
    fig,ax = plt.subplots(rows,2,figsize=(15,10))
    
    ax = ax.ravel()

    for i, column in enumerate(list_of_inputs):
    
        sns.distplot(df[column],ax=ax[i],bins=10)

        ax[i].set_xlabel(str(column))
        ax[i].set_ylabel(y_label_input)
        
        mean = np.mean(df[column])
        ax[i].set_title(title_input + ' ' +column + f' (Mean: {round(mean,2)})')
        ax[i].axvline(mean,color = 'black')

    fig.tight_layout(h_pad = 1,w_pad = 10)
```

```python
dist_plot(exams_df,['act_participation_2017','act_participation_2018','sat_participation_2017','sat_participation_2018'],
                   'Density','Distplot for')
```

## <span style="color:green"> My Analysis:
    
We see that it is difficult to label a scatter plot as there are too many values. Moreover, it would be good if we could numerically identify the top underperforming states. Hence, we shall model the linear regression ourself and identify the states that underperform according to the model.

```python
# Function to plot OLS and predict scores that the states should achieve according to their participation rates

import statsmodels.api as sm

def ols_plot(df,list_x,list_y):
    for i,_ in enumerate(list_x):

        dep_var = df[list_y[i]]
        indep_var = df[list_x[i]]
        indep_var_const = sm.add_constant(indep_var)

        model = sm.OLS(dep_var,indep_var_const)
        result = model.fit()
        print(result.summary())

        my_plot = plt.figure()
        my_plot = plt.plot(indep_var,dep_var,'ro')
        my_plot = plt.plot(indep_var,result.fittedvalues)

        plt.title(f'Linear Regression {list_y[i]} on {list_x[i]}')
        plt.ylabel('Exam Scores')
        plt.xlabel('Participation Rates')
        
        col = list_x[i]+'_predict'

        exams_df[col] = result.fittedvalues
```

```python
ols_plot(exams_df,['act_participation_2017','act_participation_2018','sat_participation_2017','sat_participation_2018'],
             ['act_composite_2017','act_composite_2018','sat_total_2017','sat_total_2018'])
```

```python
#creating a new df to store difference in the predictions

prediction_df = pd.DataFrame({'state':exams_df['state']})

prediction_df['sat_2017_diff'] = exams_df['sat_total_2017'] - exams_df['sat_participation_2017_predict']
prediction_df['sat_2018_diff'] = exams_df['sat_total_2018'] - exams_df['sat_participation_2018_predict']
prediction_df['act_2017_diff'] = exams_df['act_composite_2017'] - exams_df['act_participation_2017_predict']
prediction_df['act_2018_diff'] = exams_df['act_composite_2018'] - exams_df['act_participation_2018_predict']

prediction_df.head()
```
