## Problem Statement

The dataset used for this project is based off the greater region of Ames, Iowa. It is collected from 2006 to 2010 by the Ames Assessor's Office. The problem that I would like to tackle in this project is: 

* What are the features of a property that sellers should look out for in order to sell their houses at a higher price in Ames?

Conversely, this will also be useful for buyers who might be looking to resell their house.

Based on this problem statement, I will create a linear predictive model using using Linear, Lasso, Ridge or ElasticNet regression to investigate and identify the features which have a strong positive or negative relationship with housing sale prices.

### Analysis

Since the values of lot configuration for 'frontage on 2 sides' and 'frontage on 3 sides' (which have ftontage sizes) is present for some values but the corresponding lot frontage value is missing, we will have to estimate the numbers for those lot frontages

```python
# we first have to pull out the given mean and sd of lot frontage

# then we set those without frontage to 0

mean = train_df['lot_frontage'].mean()

sd = train_df['lot_frontage'].std()
```

```python
# Creating a random number for houses with lot frontage taking into account its mean and sd
# I tried my hand at creating my own random function here
# Alternatively, can use np.random.normal(train_df['Lot Frontage'].mean(),train_df['Lot Frontage'].std())

random.seed(1)

def random_sd(row,mean,sd):
    
    if np.isnan(row):
        rand_num = random.random()

        if rand_num<=0.34:
            result = random.uniform(mean,mean-sd)
            if result<=0:
                result = random.uniform(mean,mean-sd)

        if rand_num>0.34 and rand_num<=0.68:
            result = random.uniform(mean,mean+sd)
            if result<=0:
                result = random.uniform(mean,mean-sd)

        if rand_num> 0.68 and rand_num <=0.81:
            result = random.uniform(mean-sd,mean-2*sd)
            if result<=0:
                result = random.uniform(mean,mean-sd)

        if rand_num>0.81 and rand_num<=0.95:
            result = random.uniform(mean+sd,mean+2*sd)
            if result<=0:
                result = random.uniform(mean,mean-sd)

        if rand_num>0.95 and rand_num<=0.97:
            result = random.uniform(mean-2*sd,mean-3*sd)
            if result<=0:
                result = random.uniform(mean,mean-sd)

        if rand_num>0.97 and rand_num<=0.99:
            result = random.uniform(mean+2*sd,mean+3*sd)
            if result<=0:
                result = random.uniform(mean,mean-sd)

        if rand_num>0.99 and rand_num<=0.995:
            result = random.uniform(mean-3*sd,mean-4*sd)
            if result<=0:
                result = random.uniform(mean,mean-sd)
        else:
            result = random.uniform(mean+3*sd,mean+4*sd)
            if result<=0:
                result = random.uniform(mean+3*sd,mean+4*sd)
        
        return (result if result>0 else random_sd(row,mean,sd))
    
    else:
        return row
```

```python
train_df['lot_frontage'] = train_df['lot_frontage'].apply(
    random_sd,args = (mean,sd))

test_df['lot_frontage'] = test_df['lot_frontage'].apply(
    random_sd,args = (mean,sd))
```

```python
#Now that we have settled all the na values, let's check if any columns have conflicting dtypes

def dtype_check(df):
    
    diff_type_list = []
    
    for column in df.columns:

        d_type = df[column].dtypes

        if d_type == 'O':
            if np.array_equal(df[column], df[column].astype('str')):
                pass
            else: 
                diff_type_list.append(column)
        if d_type == 'int64':
            if np.array_equal(df[column], df[column].astype('int')):
                pass
            else: 
                diff_type_list.append(column)
        if d_type == 'float':
            if np.array_equal(df[column], df[column].astype('float')):
                pass
            else:
                diff_type_list.append(column)

    print(diff_type_list)
```
